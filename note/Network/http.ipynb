# 网络编程基础——http协议与web服务器

## 一、HTTP协议介绍

HTTP协议翻译为超文本（超越文本限制：音乐、视频、图片、超链接等）传输协议（数据格式的约定）。

设计之初是为了传输网页数据，现在允许传输任意类型数据。

HTTP协议，基于TCP协议，因此需要先建立连接。

作用：规定了浏览器和web服务器通信数据的格式，即浏览器和web服务器间的通讯需要基于http协议。

浏览器访问web服务器的流程：

在服务器的主机上运行着web服务器程序（简称web服务器）。

浏览器输入网址，经过DNS（域名解析服务器）进行域名解析（网络通讯没有域名的概念，只能通过IP）得到IP，http默认端口号是80，就可以与web服务器程序建立连接。

随后，浏览器发送http请求数据，web服务器根据请求在本地服务器找到指定资源，并以http形式发送回浏览器。

可以把HTTP理解为一种数据格式。

## 二、URL

翻译为统一资源定位符，通俗理解就是网络资源地址（网址）。

如某个网址：

https://zh.wikipedia.org/wiki/%E5%90%8C%E6%97%B6%E5%AE%9A%E4%BD%8D%E4%B8%8E%E5%9C%B0%E5%9B%BE%E6%9E%84%E5%BB%BA

由三个部分组成：

- 协议部分：https:// （数据进行加密，端口号443）， http:// （明文，端口号80），ftp:// 

- 域名部分： zh.wikipedia.org/ 对应得到ip地址  点进行分割，英文字母和数字组成

- 资源路径部分：后面的内容

有些网址后还有查询参数部分：

如，?page=1&count=10 （第1页的10条数据） 后面的参数都与之前的以&分割

## 四、HTTP请求报文

请求报文是浏览器是发送给web服务器程序的http协议数据。

有两种常用的方式：

- GET：获取web服务器数据，如获取新闻列表数据

## 五、HTTP响应报文

响应报文是web服务器程序发送给浏览器的。

如响应报头

```
——————响应（状态）行————————
HTTP/1.1（HTTP版本） 200（状态码） OK（状态码的描述）
——————响应头——————————
Date: Sun, 27 Nov 2022 10:06:09 GMT  （服务器时间，格林威治时间）
Content-Type: text/html（内容类型及编码格式）
Transfer-Encoding: chunked  （服务器发送给客户端的数据不确定长度）
（存在结束表示0\r\n，或者有一个Content—Length：200，及服务器发送200个字节）
（以上两种选项只能二选一）
Connection: keep-alive（和客户端保持长连接）
Server（服务器名称）: nginx
—————— 以下内容都是自定义响应头信息（请求头也可以自定义） 如is_login
Vary: Accept-Encoding
Access-Control-Allow-Credentials: true
Access-Control-Allow-Methods: GET, POST, PUT, DELETE
Access-Control-Allow-Headers: *
Content-Encoding: gzip
——————————空行——————————
\r\n
——————————响应体————————
真正意义上给浏览器解析的数据
```

真正解析的数据在响应体中，响应的也是每一块后都有\r\n.

## 六、状态码

三位数字，用于表示web服务器的响应状态的代码。

200：成功；307：重定向；400：错误的请求，请求资源路径地址或参数有误；404：请求资源在服务器中不存在；500，服务器内部代码出现错误。

307，如使用http://www.baidu.com/ 会重定向到https://www.baidu.com/ 状态就为200了。
